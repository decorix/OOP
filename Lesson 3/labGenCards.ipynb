```python
def cards(count_cards=36, joker="no") -> list:
    '''Функция возвращает колоду игральных карт в виде списка кортежей'''
    suit = ["черви", "крести", "бубны", "пики"]
    type_cards = ["туз", "король", "королева", "валет"]

    if (count_cards % 4 == 0):
        all_cards = list()
        type = type_cards.copy()
        count_type_cards = count_cards // len(suit)

        if (count_type_cards - len(type)) > 10:
            number = count_type_cards - len(type)
        else:
            number = 10

        while len(type) != count_type_cards:
            type.append(str(number))
            number -= 1

        for i in suit:
            one_type_cards = []
            for j in type:
                card = j + " " + i
                one_type_cards.append(card)
            all_cards.append(tuple(one_type_cards))
        
        if (joker == "yes"):
            all_cards.append(("джокер черный", "джокер красный"))

        return all_cards
    else:
        print("Количество карт должно быть кратно 4")
        return list()
    
```

```python
print(cards())
print(cards(52, 'yes'))
print(cards(64, 'yes'))
```
