```python
p = ' '

for i in range(1, 256):
    l = 4 - len(str(i)) #отступ в ячейке перед числом
    print(f"\033[48;5;{i}m{p * l}{i}\033[0m", end="")
    if (i % 16 == 0):
        print()

print()

for i in range(1, 256):
    l = 4 - len(str(i))
    print(f"\033[38;5;{i}m{p * l}{i}\033[0m", end="")
    if (i % 16 == 0):
        print()
```
