```python
# создать csv-файл, где будут храниться данные о городах, в названии которых есть буква 'ы'       
with open("C:\\Users\\User\\Desktop\\4sem\\OOP\\Lesson_5\\spisok_gorodov.csv", "r", encoding='UTF-8') as f:
    cities = []
    headers = []
    read = csv.DictReader(f, delimiter=";")
    headers = read.fieldnames
    for line in read:
        if 'ы' in line['Город']:
            cities.append(line)
            
    print(cities)
    print(headers)
    
with open(f"C:\\Users\\User\\Desktop\\4sem\\OOP\\Lesson_5\\citiesUseЫ.csv", "w") as f:
    writer = csv.DictWriter(f,lineterminator="\n", delimiter=";", fieldnames=headers)
    writer.writeheader()
    writer.writerows(cities)
    
```

```python
# создать csv-файл, где данные будут отсортированы по названию округа, региона, города
with open("C:\\Users\\User\\Desktop\\4sem\\OOP\\Lesson_5\\spisok_gorodov.csv", "r", encoding='UTF-8') as f:
    cities = []
    headers = []
    read = csv.DictReader(f, delimiter=";")
    headers = read.fieldnames
    for line in read:
        cities.append(line)
    
with open(f"C:\\Users\\User\\Desktop\\4sem\\OOP\\Lesson_5\\citiesSortOblRegCity.csv", "w") as f:
    writer = csv.DictWriter(f,lineterminator="\n", delimiter=";", fieldnames=headers)
    writer.writeheader()
    temp_data_sorted = sorted(cities, key=lambda x: (x["Федеральный округ"], x["Регион"], x["Город"]))
    writer.writerows(temp_data_sorted)
```

```python
# вывести в консоль список городов, название которых содержит до 4-х букв включительно и отсортированные в обратном алфавитном порядке
with open("C:\\Users\\User\\Desktop\\4sem\\OOP\\Lesson_5\\spisok_gorodov.csv", "r", encoding='UTF-8') as f:
    cities = set()
    read = csv.DictReader(f, delimiter=';')
    for line in read:
        if len(line['Город'])<=4:
            cities.add(line['Город'])
    
    for city in sorted(cities, reverse=True):
        print(city)
```

```python
# вывести в консоль буквы алфавита по убыванию и рядом количество городов, начинающихся на каждую букву
alp = "АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ"
alp_dict = dict()

for i in alp:
    alp_dict[i] = 0

with open("C:\\Users\\User\\Desktop\\4sem\\OOP\\Lesson_5\\spisok_gorodov.csv", "r", encoding='UTF-8') as f:
    read = csv.DictReader(f, delimiter=";")
    for line in read:
        alp_dict[line["Город"][0]] += 1

for index, value in alp_dict.items():
    print(f"{index}: {value}")
```

```python
# вывести в консоль числа по возрастанию и рядом количество городов, состоящих из указанного количества букв
cities_len = dict()
count_letters = dict()

with open("C:\\Users\\User\\Desktop\\4sem\\OOP\\Lesson_5\\spisok_gorodov.csv", "r", encoding='UTF-8') as f:
    read = csv.DictReader(f, delimiter=";")
    for line in read:
        cities_len[line["Город"]] = len(line["Город"])

for index, value in cities_len.items():
    count_letters[value] = count_letters.get(value, 0) + 1

for index, value in sorted(count_letters.items(), key=lambda para: para[0]):
    print(f"{index}: {value}")
```

```python
# вывести в консоль список ТОП-10 городов с самыми длинными названиями
cities_len = dict()
count = 0

with open("C:\\Users\\User\\Desktop\\4sem\\OOP\\Lesson_5\\spisok_gorodov.csv", "r", encoding='UTF-8') as f:
    read = csv.DictReader(f, delimiter=";")
    for line in read:
        cities_len[line["Город"]] = len(line["Город"])

for i in sorted(cities_len.items(), key=lambda para: para[1], reverse=True):
    print(i[0])
    
    count += 1

    if (count == 10):
        break
```

```python
# вывести в консоль список ТОП-10 городов, в названии которых больше всего согласных букв
consonants = "бвгджзйклмнпрстфхцчшщ"
cities_count = dict()
count = 0

with open("C:\\Users\\User\\Desktop\\4sem\\OOP\\Lesson_5\\spisok_gorodov.csv", "r", encoding='UTF-8') as f:
    read = csv.DictReader(f, delimiter=";")
    for line in read:
        cities_count[line["Город"]] = 0
        for i in line["Город"].lower():
            if i in consonants:
                cities_count[line["Город"]] += 1

for i in sorted(cities_count.items(), key=lambda para: para[1], reverse=True):
    print(i[0])

    count += 1

    if (count == 10):
        break
    
```
