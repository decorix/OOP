```python
import sqlite3
from prettytable import from_db_cursor
import json
```

```python
db_name = "mosmetro.db"
```

```python
db = sqlite3.connect(db_name)
cur = db.cursor()
```

```python
cur.execute("CREATE TABLE line (id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL, name VARCHAR(100) UNIQUE)")
db.commit()
```

```python
cur.execute("CREATE TABLE admarea (id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL, name VARCHAR(100) UNIQUE)")
db.commit()
```

```python
cur.execute("CREATE TABLE station (id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL, id_station INTEGERE NOT NULL, name VARCHAR(100) NOT NULL, id_line INTEGER NOT NULL, id_admarea INTEGER NOT NULL, status VARCHAR(100) NOT NULL, id_global INTEGER NIT NULL, CONSTRAINT 'FK_station_line' FOREIGN KEY ('id_line') REFERENCES 'line' ('id'), CONSTRAINT 'FK_station_admarea' FOREIGN KEY ('id_admarea') REFERENCES 'admarea' ('id'))")
db.commit()
```

```python
stations = list()
lines = set()
admareas = set()
order_lines = dict()
order_admareas = dict()
count = 1

with open("C:\\Users\\User\\Desktop\\4sem\\OOP\\Lesson_6\\mosmetro.json", "r", encoding='UTF-8') as f:
    d = json.load(f)

for station in d:
    stations.append(station)
    lines.add(station["Line"])
    admareas.add(station["AdmArea"])

for line in lines:
    cur.execute("INSERT INTO 'line' ('name') VALUES (?)", (line,))
    order_lines[line] = count
    count += 1

count = 1

for admarea in admareas:
    cur.execute("INSERT INTO 'admarea' ('name') VALUES (?)", (admarea,))
    order_admareas[admarea] = count
    count += 1

for station in stations:
    cur.execute("INSERT INTO 'station' ('id_station', 'name', 'id_line', 'id_admarea', 'status', 'id_global') VALUES (?, ?, ?, ?, ?, ?)", (station["ID"], station["Station"], order_lines[station["Line"]], order_admareas[station["AdmArea"]], station["ObjectStatus"], station["global_id"]))

db.commit()
```

```python
# вывести всю информацию о станциях, которые расположены в округе чье имя начинается на Н
cur.execute("SELECT station.id, station.id_station, station.name, line.name as line_name, admarea.name as admarea_name, station.status, station.id_global FROM station JOIN line JOIN admarea ON line.id = station.id_line AND admarea.id = station.id_admarea WHERE admarea.name LIKE 'Н%';")
from_db_cursor(cur)
```

```python
# вывести всю информацию о станциях, которые расположены в округе чье имя начинается на З или расположенные на Кольцевой линии
cur.execute("SELECT station.id, station.id_station, station.name, line.name as line_name, admarea.name as admarea_name, station.status, station.id_global FROM station JOIN line JOIN admarea ON line.id = station.id_line AND admarea.id = station.id_admarea WHERE admarea.name LIKE 'З%' OR line.name = 'Кольцевая линия';")
from_db_cursor(cur)
```
